В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

Описание данных
* Name — название игры
* Platform — платформа

* EU_sales — продажи в Европе (миллионы проданных копий)
* JP_sales — продажи в Японии (миллионы проданных копий)

* Critic_Score — оценка критиков (максимум 100)
* User_Score — оценка пользователей (максимум 10)
* Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

```python
df.isnull().sum()
```

Обратим внимание на аббревиатуру 'tbd' в столбце с оценкой пользователей. TBD - аббревиатура от английского To Be Determined (будет определено) или To Be Decided (будет решено), значит значения 'tbd' можно заменить на NaN. 

```python
df['user_score'] = df['user_score'].replace('tbd', np.nan) 
df['user_score'] = df['user_score'].astype(float) # object преобразуем в float64, т.к. оценка имеет не целочисленное значение
df.info()
```

```python
# При необходимости обработаем пропуски;
# Пропуски в стобце name удалим, т.к. не удастся определить что это за игры
df.dropna(subset=['name'], inplace=True)
df.isnull().sum()
```

**Объясним, почему заполнили пропуски определённым образом или почему не стали это делать.**
    Пропуски в оставшихся столбцах ('critic_score','user_score','rating') заполнять медианными или средними значениями по жанру не имеет смысла, т.к. пропущенных значений очень много и заполненные пропуски смогут повлиять на ход дальнейшего исследования.
    Удалять строки с пропущенными значениями в этих столбцах тоже не стоит, т.к. полезные для исследования данные могут содержаться в других столбцах.

Опишем причины, которые могли привести к пропускам:

- данные в столбце name и genre могут быть пропущены в связи с техническим сбоем, или сопряжены с человеческим фактором;
- данные в столбце year_of_release могут быть пропущены в связи с допущенной ошибкой человека при внесении в базу данных информации;
- данные в столбце critic_score и user_score могут отсутствовать по причине того, что в период, когда собирались данные эти игры уже ушли с рынка и не были оценены или же оценка просто не проводилась среди пользователей и экспертов;
- для определения рейтинга (столбец raiting) издатель присылает в ESRB видеоролик, который содержит наиболее впечатляющие и резкие сцены игры. Также издатель заполняет специальную анкету сведениями о содержании игры. Много пропущенных значений в этом столбце может означать, что издатель не направлял вышеуказанную информацию в ESRB.

- Посмотрим, как менялись продажи по платформам. Выберем платформы с наибольшими суммарными продажами и построим распределение по годам, ответим на вопрос за какой характерный срок появляются новые и исчезают старые платформы?

```python
df.groupby('platform')['all_sales'].sum().sort_values()
```

```python
#большими продажами будем считать продажи свыше 300 млн.экземпляров
df_new_top_sales = df.groupby('platform').filter(lambda x: x['all_sales'].sum() > 300)
```

```python
pivot_all_sales = df_new_top_sales.pivot_table(index=['platform', 'year_of_release'], values='all_sales', aggfunc='sum')
pivot_all_sales
```

```python
# применяем метод unstack() к сводной таблице
sales_by_platform_year = pivot_all_sales.unstack(level=0)

# строим распределение продаж платформ по годам
sales_by_platform_year.plot(kind='bar', figsize=(18,4), width=1.3, title='Продажи платформ по годам')
plt.xlabel('Год выпуска игры')
plt.ylabel('Продажи, млн. копий')
plt.show()
```

**Популярные игровые платформы с общими продажами игр более 300 млн. экземпляров появляются на рынке каждые 1-5 лет, а исчезают игровые платформы с рынка за 5-11 лет.**

- Посмотрим, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков. Построим диаграмму рассеяния и посчитаем корреляцию между отзывами и продажами. Сформулируем выводы.

```python
# Выбираем строки с платформой X360
df_subset_x360 = df_subset[df_subset.reset_index()['platform'] == 'X360']
```

#### Шаг 5. Проверка гипотез

1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
2. Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Самостоятельно зададим пороговое значение alpha.

Поясним:
- Как мы сформулировали нулевую и альтернативную гипотезы;
- Какой критерий применили для проверки гипотез и почему.

```python
'''
В Python есть специальный метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей
по взятым из них выборкам: ttest_ind. 

Три условия для использования этого метода выполняются:
1. Генеральные совокупности не зависят друг от друга.
2. Выборочные средние нормально распределены.
3. Дисперсии рассматриваемых генеральных совокупностей равны.
'''
#Ho = cредние пользовательские рейтинги платформ Xbox One и PC равны;
#H1 = cредние пользовательские рейтинги платформ Xbox One и PC не равны.
```

```python
#отфильтруем датафрейм с данными за последние 5 лет
xone = df[(df['platform'] == 'XOne') & (df['user_score'] > 0)]['user_score']
pc = df[(df['platform'] == 'PC') & (df['user_score'] > 0)]['user_score']
```

```python
#посчитаем средние пользовательские рейтинги этих платформ
print(f'Среднее значение пользовательского рейтинга для XOne: {xone.mean()}')
print(f'Среднее значение пользовательского рейтинга для PC: {pc.mean()}')
```

```python
# задаём уровень значимости
alpha = 0.05

# проводим ttest для двух выборок, в качестве альтернативной используем двустороннюю гипотезу "не равна"
results = st.ttest_ind(xone, pc, alternative='two-sided')

print(f'p-value: {results.pvalue}')

# проверяем p-value
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Нет оснований отвергнуть нулевую гипотезу')
```

**Полученное значение p-value (~0.62676) больше заданного уровня значимости, поэтому нельзя отвергнуть нулевую гипотезу. Есть основания говорить, что пользовательские рейтинги XOne и PC равны.**

Аналогично проверим гипотизу о том, что средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

```python
#отфильтруем датафрейм с данными за последние 5 лет
action = df[(df['genre'] == 'Action') & (df['user_score'] > 0)]['user_score']
sports = df[(df['genre'] == 'Sports') & (df['user_score'] > 0)]['user_score']
```

```python
#посчитаем средние пользовательские рейтинги этих платформ
print(f'Среднее значение пользовательского рейтинга для жанра Action: {action.mean()}')
print(f'Среднее значение пользовательского рейтинга для жанра Sports: {sports.mean()}')
```

```python
#Ho = cредние пользовательские рейтинги Аction и Sports равны;
#H1 = cредние пользовательские рейтинги жанра Action больше.

# задаём уровень значимости
alpha = 0.05

# проводим ttest для двух выборок, в качестве альтернативной используем правостороннюю гипотезу
results = st.ttest_ind(action, sports, alternative='greater')

print(f'p-value: {results.pvalue}')

# проверяем p-value
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Нет оснований отвергнуть нулевую гипотезу')
```

**Полученное значение p-value (~3.35) меньше заданного уровня значимости, поэтому можно отвергнуть нулевую гипотезу. Есть основания полагать, что пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.**
