**Стадии:**
1. Флотация
    Во флотационную установку подаётся смесь золотосодержащей руды. После обогащения получается черновой концентрат и «отвальные хвосты», то есть остатки продукта с низкой концентрацией ценных металлов.
    На стабильность этого процесса влияет непостоянное и неоптимальное физико-химическое состояние флотационной пульпы (смеси твёрдых частиц и жидкости).
2. Очистка 
    Черновой концентрат проходит две очистки. На выходе получается финальный концентрат и новые отвальные хвосты.

**Описание данных:**
- Технологический процесс
- Rougher feed — исходное сырье
- Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
- Xanthate — ксантогенат (промотер, или активатор флотации);
- Sulphate — сульфат (на данном производстве сульфид натрия);
- Depressant — депрессант (силикат натрия).
- Rougher process (англ. «грубый процесс») — флотация
- Rougher tails — отвальные хвосты
- Float banks — флотационная установка
- Cleaner process — очистка
- Rougher Au — черновой концентрат золота
- Final Au — финальный концентрат золота
- Параметры этапов
- air amount — объём воздуха
- fluid levels — уровень жидкости
- feed size — размер гранул сырья
- feed rate — скорость подачи

**Наименование признаков**

Наименование признаков должно быть такое:
[этап].[тип_параметра].[название_параметра]
Пример: rougher.input.feed_ag

   - Возможные значения для блока [этап]:
     - rougher — флотация
     - primary_cleaner — первичная очистка
     - secondary_cleaner — вторичная очистка
     - final — финальные характеристики
   - Возможные значения для блока [тип_параметра]:
     - input — параметры сырья
     - output — параметры продукта
     - state — параметры, характеризующие текущее состояние этапа
     - calculation — расчётные характеристики

### Проверим, что эффективность обогащения рассчитана правильно.

<div class="alert-info">
Для расчета эффективности (recovery = ((C*(F-T)) / (F*(C-T)))*100)
обогащения необходимо использовать переменные доступные в обучающем датасете:

C — доля золота в концентрате после флотации/очистки 'rougher.output.concentrate_au';
    
F — доля золота в сырье/концентрате до флотации/очистки 'rougher.input.feed_au';
    
T — доля золота в отвальных хвостах после флотации/очистки 'rougher.output.tail_au'.
    
Для прогноза коэффициента нужно найти долю золота в концентратах
</div>

```python
#Вычислим эффективность на обучающей выборке для признака rougher.output.recovery.
def recovery(df):
    recovery = (((df['rougher.output.concentrate_au']*(df['rougher.input.feed_au']-df['rougher.output.tail_au'])) / \
    (df['rougher.input.feed_au']*(df['rougher.output.concentrate_au']-df['rougher.output.tail_au'])))*100)
    return recovery
```

```python
recovery_calc = recovery(df_train)
print(f'Эффективность обогащения руды:\n{recovery_calc}')
```

```python
#Найдем MAE между расчётами значениями и значением признака.
#значения признака хранятся в переменной 'rougher.output.recovery' тренировочной выборки
input_recovery = df_train['rougher.output.recovery']
print('Средняя абсолютная ошибка =', mean_absolute_error(input_recovery, recovery_calc))
```

- эффективность обогащения чернового концентрата - rougher.output.recovery;
- эффективность обогащения финального концентрата - final.output.recovery.

### Напишем функцию для вычисления итоговой sMAPE.

Формулы метрик качества:



```python
def smape(target, pred):
    smape_nom = abs(target-pred)/((abs(target)+abs(pred)/2))*100
    smape_nom = smape_nom.fillna(value=0)
    smape_nom = np.mean(smape_nom)
    return smape_nom
```

![Итоговое sMAPE](https://pictures.s3.yandex.net/resources/_smape_1576239054.jpg "Итоговое sMAPE")

```python
#посчитаем итоговый smape
def smape_sum(target_rougher, pred_rougher, target_final, pred_final):
    smape_rougher = smape(target_rougher, pred_rougher)
    smape_final = smape(target_final, pred_final)
    #итоговый smape
    smape_sum_total = 0.25 * smape_rougher + 0.75 * smape_final
    
    return smape_sum_total
```

```python
%%time
#Линейная регрессия
smape_scorer = make_scorer(smape, greater_is_better=False) #создадим объект-оценщик, для нашей модели
model_regression = LinearRegression() #создаем объект линейной регрессии


#оценим модель с помощью кроссвалидации для rougher.output.recovery
model_regression_scores_rougher = cross_val_score(model_regression,
                                          features_train,
                                          target_train['rougher.output.recovery'],
#рассчитаем симметричное среднее абсолютное процентное отклонение
                                          scoring = smape_scorer, 
                                          cv = 5)
smape_rougher = np.mean(abs(model_regression_scores_rougher))
print(f'Средняя оценка качества модели (sMAPE) по rougher.output.recovery - \
симметричное среднее абсолютное процентное отклонение: {round(smape_rougher, 2)} %')


#оценим модель с помощью кроссвалидации для final.output.recovery
model_regression_scores_final = cross_val_score(model_regression,
                                          features_train,
                                          target_train['final.output.recovery'],
#рассчитаем симметричное среднее абсолютное процентное отклонение
                                          scoring = smape_scorer, 
                                          cv = 5)
smape_final = np.mean(abs(model_regression_scores_final))
print(f'Средняя оценка качества модели (sMAPE) по final.output.recovery - \
симметричное среднее абсолютное процентное отклонение: {round(smape_final, 2)} %')
```

```python
#итоговый sMAPE
smape_sum = 0.25 * smape_rougher + 0.75 * smape_final
print(f'Итоговая средняя оценка sMAPE {round(smape_sum, 2)} %')
```

```python
%%time

#Случайный лес
model_rnd_forest = RandomForestRegressor(random_state=12345)

#подберем оптимальные параметры глубины для дерева
parametrs_tree = {'max_depth': [x for x in range(1, 15, 3)],
                  'n_estimators': [x for x in range(10, 110, 50)]
                 }

grid_rnd_forest_rougher = GridSearchCV(model_rnd_forest,
                                       parametrs_tree,
                                       cv = 5,
                                       scoring = smape_scorer)

#обучим модель
grid_rnd_forest_rougher.fit(features_train, target_train['rougher.output.recovery'])
print(f'Лучшие параметры модели grid_rnd_forest_rougher: {grid_rnd_forest_rougher.best_params_}')
print(f'Оценка качества модели (sMAPE) - симметричное среднее абсолютное процентное отклонение: \
{round(abs(grid_rnd_forest_rougher.best_score_), 2)} %')   

print()

grid_rnd_forest_final = GridSearchCV(model_rnd_forest,
                                       parametrs_tree,
                                       cv = 5,
                                       scoring = smape_scorer)

#обучим модель
grid_rnd_forest_final.fit(features_train, target_train['final.output.recovery'])  
    
print(f'Лучшие параметры модели grid_rnd_forest_final: {grid_rnd_forest_final.best_params_}')
print(f'Оценка качества модели (sMAPE) - симметричное среднее абсолютное процентное отклонение: \
{round(abs(grid_rnd_forest_final.best_score_), 2)} %')    
```

```python
#итоговый sMAPE
smape_sum_rnd_forest = 0.25 * round(abs(grid_rnd_forest_rougher.best_score_), 2) \
+ 0.75 * round(abs(grid_rnd_forest_final.best_score_), 2)
print(f'Итоговая средняя оценка sMAPE {round(smape_sum_rnd_forest, 2)} %')
```

<div class="alert-info">Написанная нами модель случайного леса на кросс-валидации показала лучшее итоговое среднее значение sMAPE - 5.22% среди всех моделей, при глубине дерева - 4 и числе деревьев - 10 для модели grid_rnd_forest_rougher и при глубине дерева - 4 и числе деревьев - 60 для модели grid_rnd_forest_final. Используем её для предсказания ['rougher.output.recovery','final.output.recovery'] на тестовой выборке.</div>
